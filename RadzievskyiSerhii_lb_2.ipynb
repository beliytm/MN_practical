```python
#Зчитайте дані з таблиці "Середньорічна щомісячна зарплата в Україні за 1992—2022 роки" з Вікіпедії за посиланням  та виконайте наступні завдання:

import pandas as pd
import numpy as np


url = "https://uk.wikipedia.org/wiki/%D0%A1%D0%B5%D1%80%D0%B5%D0%B4%D0%BD%D1%8F_%D0%B7%D0%B0%D1%80%D0%BE%D0%B1%D1%96%D1%82%D0%BD%D0%B0_%D0%BF%D0%BB%D0%B0%D1%82%D0%B0_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96"

tables = pd.read_html(url, header=0)

# Обираємо потрібну таблицю (можливо, знадобиться вказати індекс)
df = tables[2]
```

```python
#2.	Визначте кількість рядків і стовпчиків за допомогою атрибуту shape

rows, columns = df.shape
print(f"Кількість рядків: {rows}, Кількість стовпчиків: {columns}")
```

```python
#3.	Визначте загальну кількість років та кількість місяців у даних
years = len(df['Рік'])
months = years*12
print(f"Загальна кількість років: {years}, Загальна кількість місяців: {months}")
```

```python
#4.	Замінити пропущені значення  «-» в таблиці на NaN
df = df.replace('—', np.nan)
df
```

```python
#5. Визначити тип стовпців за допомогою dtypes.
col = df.dtypes
print("Типи стовпців:")
print(col)
```

```python
#6.	Видаліть передостанній і останній стовпчики ( Кількість кгхліба[20] на середнюзарплату, 
#Кільксть БігМаків на середню зарплату). Далі працюйте з оновленою таблицею. 
df = df.drop(columns=["Кількість кг хліба[20] на середню зарплату", "Кількість БігМаків на середню зарплату"])
df
```

```python
#7.	Замінити тип нечислових даних на числові.
#Видалимо * в 2022* і 2023*
df["Рік"] = df["Рік"].str.replace("*", "")
df["Рік"] = df["Рік"].astype(int)
df
```

```python
df["В національній валюті"] = df["В національній валюті"].str.replace('▲', "").str.replace('▼', "")
df["В доларах США"] = df["В доларах США"].str.replace('▲', "").str.replace('▼', "")

df['В національній валюті'] = df['В національній валюті'].str.replace(',', '.').astype(float)
df['В доларах США'] = df['В доларах США'].str.replace(',', '.').astype(float)
df

#df = df.apply(pd.to_numeric, errors='coerce')
```

```python
#8.Порахуйте частку пропусків в кожній колонці (за допомогою методів isnull , sum, len)

missing_percentage = df.isnull().sum() / len(df) * 100

print("Частка пропусків в кожній колонці:")
print(missing_percentage)
```

```python
#9.Замініть всі пропущені значення на середні значення по (стовпцю).
df1 = df.fillna(df.mean())
print(df1)
```

```python
#10.Знайдіть рік з найвищою зарплатою в Україні в грн і вкажіть цю зарплату
# Знайдемо індекс рядка з максимальною вартістю в національній валюті
index_max_value = df['В національній валюті'].idxmax()

# Отримаємо відповідну максимальну вартість
max_value = df.loc[index_max_value, 'В національній валюті']

# Визначимо рік, який відповідає цій максимальній вартості
year = df.loc[index_max_value, 'Рік']

# Виведення результату
print("Рік з найвищою вартістю в національній валюті:", year, "\nЗарплата", max_value)
```

```python
#11.Знайдіть рік з мінімальною  зарплатою в Україні в дол. США і вкажіть цю зарплату.

index_min_value = df['В доларах США'].idxmin()
min_value = df.loc[index_min_value, 'В доларах США']

# Визначимо рік, який відповідає цій максимальній вартості
year_ = df.loc[index_min_value, 'Рік']

# Виведення результату
print("Рік з мінімальною  зарплатою в Україні в дол. США:", year_, "\nЗарплата", min_value)
```

```python
#12.	Виведіть роки, коли зарплата в грн була вище середньої.
# Знайдемо середню вартість зарплати в національній валюті
average_salary = df['В національній валюті'].mean()
print("Середня зарплата == ", average_salary)

# Виберемо роки, коли зарплата була вище середньої
years_above = df[df['В національній валюті'] > average_salary]['Рік']

print("Роки, коли зарплата в В національній валюті була вище середньої:")
print(years_above.to_string(index=False))
```

```python
#13.Порівняйте  зарплату  в грн за останні два доступних роки.  Знайдемо дані для років 2022 і 2023
year_2022 = df[df['Рік'] == 2022]['В національній валюті'].values[0]
year_2023 = df[df['Рік'] == 2023]['В національній валюті'].values[0]

# Порівняємо значення
if year_2022 > year_2023:
    print("В 2022 році зарплата в національній валюті більше ніж в 2023 році.")
elif year_2022 < year_2023:
    print("В 2023 році зарплата в національній валюті більше ніж в 2022 році.")
else:
    print("Зарплата в національній валюті однакова в обох роках (2022 і 2023).")
```
