1.Створіть одновимірний масив (вектор) з добутками парних чисел на їхні попередні непарні числа, починаючи з 2, та виведіть його значення. 

```python
n = 10

# Створюємо вектор, використовуючи NumPy
vector = np.zeros(n)

# Заповнюємо вектор значеннями
for i in range(1, n):
    # Парні числа
    if i % 2 == 0:
        vector[i] = i * (i - 1)  # Добуток парного числа на його попереднє непарне число

# Виводимо значення вектора
print(vector)
```

2.Створіть двовимірний масив (матрицю) розміром 3x3, заповніть його значеннями квадратів відстаней між кожною парою натуральних чисел, починаючи з 1, та виведіть його значення. 

```python

# Створюємо двовимірний масив, використовуючи NumPy
matrix = np.zeros((3, 3))

# Заповнюємо матрицю значеннями квадратів відстаней між кожною парою чисел
for i in range(n):
    for j in range(n):
        # Використовуємо квадрат відстані між (i+1) та (j+1)
        matrix[i, j] = (i + 1 - (j + 1)) ** 2

# Виводимо значення матриці
print(matrix)
```

3.Створіть масив розміром 5x5, заповніть його випадковими дійсними числами від -2 до 2 та знайдіть середнє за модулем значення його елементів. 

```python
matrix = np.random.uniform(-2, 2, size=(5, 5))

# Знаходимо середнє за модулем значення елементів матриці
average_absolute_value = np.mean(np.abs(matrix))

# Виводимо матрицю та середнє за модулем значення елементів
print("Матриця:")
print(matrix)
print("\nСереднє за модулем значення елементів:", average_absolute_value)
```

4.Створіть масив розміром 5x5, заповніть його випадковими дійсними числами в діапазоні від 0 до 1 та обчисліть суму елементів, які менше середнього значення. 

```python

matrix = np.random.rand(5, 5)

# Знаходимо середнє значення елементів матриці
average_value = np.mean(matrix)

# Знаходимо елементи, які менше середнього значення
less_than_average = matrix[matrix < average_value]

# Знаходимо суму цих елементів
sum_less_than_average = np.sum(less_than_average)

# Виводимо матрицю та результат
print("Матриця:")
print(matrix)
print("\nСереднє значення елементів:", average_value)
print("Сума елементів, які менше середнього значення:", sum_less_than_average)
```

5.Створіть два одновимірних масиви розміром 7, заповніть їх випадковими дійсними числами від -2 до 2 та виконайте на них поелементні операції додавання, віднімання ta множення. 

```python
array1 = np.random.uniform(-2, 2, 7)
array2 = np.random.uniform(-2, 2, 7)

# Виконуємо поелементні операції
addition_result = array1 + array2
subtraction_result = array1 - array2
multiplication_result = array1 * array2

# Виводимо вихідні масиви та результати операцій
print("Масив 1:", array1)
print("Масив 2:", array2)
print("\nРезультат додавання:", addition_result)
print("Результат віднімання:", subtraction_result)
print("Результат множення:", multiplication_result)
```

6.Створіть два вектори розміром 7, заповніть їх довільними цілими числами та знайдіть їх скалярний добуток.

```python
vector1 = np.random.randint(-10, 10, 7)
vector2 = np.random.randint(-10, 10, 7)

# Знаходимо скалярний добуток векторів
dot_product_result = np.dot(vector1, vector2)

# Виводимо вихідні вектори та результат скалярного добутку
print("Вектор 1:", vector1)
print("Вектор 2:", vector2)
print("\nРезультат скалярного добутку:", dot_product_result)
```

8.Створіть матрицю розміром 2x2, заповніть її випадковими цілими числами в діапазоні від 1 до 10 та знайдіть її обернену матрицю. 

```python
matrix = np.random.randint(1, 11, size=(2, 2))

# Перевіряємо, чи матриця є оберненою (детермінант не дорівнює нулю)
det = np.linalg.det(matrix)

if det != 0:
    # Знаходимо обернену матрицю
    inverse_matrix = np.linalg.inv(matrix)

    # Виводимо вихідну матрицю та обернену матрицю
    print("Матриця:")
    print(matrix)
    print("\nОбернена матриця:")
    print(inverse_matrix)
else:
    print("Матриця не має оберненої матриці (детермінант дорівнює нулю).")
```

12.Створіть дві матриці розміром 4x4, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть добуток елементів діагоналі першої матриці. 

```python
matrix1 = np.random.randint(1, 11, size=(4, 4))

# Знаходимо добуток елементів діагоналі першої матриці
diagonal_product = np.prod(np.diag(matrix1))

# Виводимо вихідну матрицю та добуток елементів діагоналі
print("Матриця 1:")
print(matrix1)
print("\nДобуток елементів діагоналі першої матриці:", diagonal_product)
```

15.Створіть матрицю розміром 5x5, заповніть її випадковими дійсними числами в діапазоні від -0.5 до 0.5 та знайдіть вектор-стовпчик, що містить кількість нульових елементів в кожному рядку матриці. 

```python
matrix = np.random.uniform(-0.5, 0.5, size=(5, 5))
# Знаходимо вектор-стовпчик кількості нульових елементів в кожному рядку матриці
zero_count_vector = np.sum(matrix == 0, axis=1).reshape(-1, 1)

# Виводимо матрицю та вектор-стовпчик
print("Матриця:")
print(matrix)
print("\nВектор-стовпчик кількості нульових елементів в кожному рядку:")
print(zero_count_vector)
```
