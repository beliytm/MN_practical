```python
#8. Побудуйте діаграму розподілу переглядів (використовуйте kind='hist'):
data['video views'].plot(kind='hist')
```

```python
#9. Визначте максимальну, мінімальну, середню кількість переглядів (використовуйте методи max, min, mean)
maxx = data['video views'].max()

minn = data['video views'].min()

meann = data['video views'].mean()

print(f"Максимальна кількість переглядів: {maxx}")
print(f"Мінімальна кількість переглядів: {minn}")
print(f"Середня кількість переглядів: {meann}")
```

```python
#10.Знайдіть країну, де найбільше відео було завантажено на YouTube
max_uploads = data[data['uploads'] == data['uploads'].max()]['Country'].values[0]
print(f"Країна з найбільшою кількістю завантажень: {max_uploads}")

```

```python
#11. Знайдіть відео (виведіть його назву) з найбільшою та найменшою кількістю завантажень uploads.
max_uploads = data[data['uploads'] == data['uploads'].max()]['Title'].values[0]
min_uploads = data[data['uploads'] == data['uploads'].min()]['Title'].values[0]
print(f"Відео з найбільшою кількістю завантажень: {max_uploads}")
print(f"Відео з найменшою кількістю завантажень: {min_uploads}")
```
